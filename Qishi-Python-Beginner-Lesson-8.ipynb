```python
# Create a new figure of size 10x8
plt.figure(figsize=(6,4))

x = np.linspace(-np.pi, np.pi, 100)
y, z = np.sin(x), np.cos(x)

# Plot sine using blue color with a continuous line of width 1 (pixels)
# plt.plot(x, y, color='blue',linewidth=1.0, linestyle='-.')
#           marker='o',ms=15,mfc='white',mec='red', alpha=0.9)
plt.plot(x, y, color='blue',linewidth=1.0, linestyle='-', label='sine')

# Plot cosine using blue color with a continuous line of width 1 (pixels)
# plt.plot(x, z, color='green',linewidth=1.0, linestyle='-')
plt.plot(x, z, color='green',linewidth=1.0, linestyle='-', label='cosine')

plt.legend(loc='upper left',frameon=True)
# Set x limits
plt.xlim(-4, 4)

# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

# Set y limits
plt.ylim(-1.0, 1.0)
#plt.ylim(-1.1, 1.1)

# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.text(-1.6, 0.8, "sin(x)")
plt.annotate(xy=(2.6,0.5), xytext=(3,0.8), s="cos(x)", arrowprops=dict(facecolor='white', shrink=0.05))
# # ax = plt.gca() # returns the current axes
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.spines['bottom'].set_position(('data',0))
# ax.yaxis.set_ticks_position('left')
# ax.spines['left'].set_position(('data',0))

plt.hlines(0, -4, 2, colors='b')
plt.axhline(0)
plt.axvline(0)

plt.show()
```

#### Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(8,6))
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# object-oriented design
fig = plt.figure(dpi=100, figsize=(8,6))

np.random.seed(1234)

# make up some data in the interval [0, 1]
y = np.random.normal(loc=0, scale=1, size=1000)
#y = y[(y > 0) & (y < 1)]
x = np.arange(len(y))

ax1 = fig.add_subplot(221)
ax1.plot(x, y, color='r')
ax1.set(title="t", xlabel="x", ylabel="y")
ax1.set_xlim(-50, 1050)
ax1.grid(True)

ax2 = fig.add_subplot(222)
ax2.hist(y)
ax2.set_xlabel('dist')
ax2.grid(True)

ax3 = fig.add_subplot(223)
ax3.boxplot(y)
ax3.set_xlabel('box')
ax3.grid(True)

ax4 = fig.add_subplot(224)
ax4.hist(y)
ax4.set_xlabel('dist')
ax4.grid(False)

fig.savefig('dist.png', dpi=300)


# fig = plt.figure(dpi=100, figsize=(8,4))
# ax1 = fig.add_subplot(221)
# ax2 = fig.add_subplot(222)
# ax3 = fig.add_subplot(223)
# ax4 = fig.add_subplot(224)
```
